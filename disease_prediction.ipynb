```python
import pandas as pd

# Load dataset.csv
df = pd.read_csv("archive/dataset.csv")

# Load Symptom-severity.csv
df_severity = pd.read_csv("archive/Symptom-severity.csv")

# Convert symptoms to lowercase for consistency
df_severity["Symptom"] = df_severity["Symptom"].str.lower()
df = df.fillna("None")  # Fill missing values in dataset

# Create a dictionary {symptom: severity}
disease_severity = dict(zip(df_severity["Symptom"], df_severity["weight"]))

# Replace symptom names with severity weights
for col in df.columns[1:]:  # Ignore 'Disease' column
    df[col] = df[col].str.lower().map(disease_severity).fillna(0).astype(int)

# Display updated dataset
df.head()
```
